```python
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vT_jChgNsQbHbg4TGepzIqk8XC9DTIKmyyxb1upo5cfZCgbfIUQc2ZC0YMzuU5uApP140Ob49KBjdqh/pub?gid=1198589591&single=true&output=csv')
df.head()
```

```python
df['datetime']= pd.to_datetime(df['date'])
df.head()
```

```python
## displaying random sample
df['date'].sample(n =10).sort_index()
```

```python
df['datetime']= pd.to_datetime(df['date'], format = '%Y%m%d')
df.head()
```

```python
df.info()
```

```python
df = df.set_index('datetime')
df
```

```python
df = df.loc['2000':, ['precipitation','mean_temp','min_temp', 'max_temp', 'snow_depth']]
df
```

```python
#missing data
df.isna().sum()
```

```python
df['mean_temp']= df['mean_temp'].interpolate()
df['min_temp']= df['min_temp'].interpolate()
df['max_temp']= df['max_temp'].interpolate()
df['precipitation'] = df['precipitation'].fillna(0)
df['snow_depth'] = df['snow_depth'].fillna(0)
```

```python
df.isna().sum()
```

```python
min_maxtemp_date = df['max_temp'].idxmin()
min_maxtemp_date
```

```python
prec1, meantemp1, mintemp1, maxtemp1, snow1 = df.loc[min_maxtemp_date]
prec1, meantemp1, mintemp1, maxtemp1, snow1
```

```python
print(f"The lowest temperature of {maxtemp1} degrees occurred on {min_maxtemp_date.strftime('%B %d,%Y')} which was a {min_maxtemp_date.day_name()}.")
```

```python
max_prec_date = df['precipitation'].idxmax()
max_prec_date
```

```python
prec2, meantemp2, mintemp2, maxtemp2, snow2 = df.loc[max_prec_date]
prec2, meantemp2, mintemp2, maxtemp2, snow2
```

```python
print(f"The highest daily precipitation of {prec2} inches occurred on {max_prec_date.strftime('%B %d,%Y')} which was a {max_prec_date.day_name()} .")
```
